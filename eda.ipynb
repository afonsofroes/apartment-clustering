```python
# load and concat jan_apr, may_june, jul_aug

jan_apr = pd.read_csv('dados_afonso/room_say_report/reservas_por_data_criada/jan_apr.csv')
may_june = pd.read_csv('dados_afonso/room_say_report/reservas_por_data_criada/may_june.csv')
jul_aug = pd.read_csv('dados_afonso/room_say_report/reservas_por_data_criada/jul_aug.csv')

desc_data = pd.concat([jan_apr, may_june, jul_aug])
```

```python
desc_data['Booked At'] = pd.to_datetime(desc_data['Booked At'], format='%Y/%m/%d %H:%M:%S')
desc_data['Check In'] = pd.to_datetime(desc_data['Check-In'], format='%Y/%m/%d')
desc_data.info()
```

```python
# remove the time from datetime object 

desc_data['Booked At'] = desc_data['Booked At'].dt.date
```

```python
desc_data['Booked At'] = pd.to_datetime(desc_data['Booked At'], format='%Y/%m/%d')
```

```python
desc_data['days_ahead'] = desc_data['Check In'] - desc_data['Booked At']
desc_data['days_ahead'] = desc_data['days_ahead'].dt.days
```

```python
desc_data['discounted'] = desc_data['Room'].str.find('%') != -1
```

```python
desc_data['discounted'] = desc_data['discounted'].astype(int)
```

```python
desc_data.head()
```

```python
desc_data = desc_data.drop(columns=['Booked At', 'Room', 'Guests', 'Check In'])
```

```python
# parse Check-In and Check-Out to datetime

desc_data['Check-In'] = pd.to_datetime(desc_data['Check-In'], format='%Y/%m/%d')
desc_data['Check-Out'] = pd.to_datetime(desc_data['Check-Out'], format='%Y/%m/%d')

desc_data['stay'] = desc_data['Check-Out'] - desc_data['Check-In']
desc_data['stay'] = desc_data['stay'].dt.days
```

```python
desc_data.head()
```

```python
disc_data = desc_data[desc_data.Room.str.find('%') != -1]
no_disc_data = desc_data[desc_data.Room.str.find('%') == -1]
```

```python
print(disc_data.shape)
print(no_disc_data.shape)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
sns.histplot(disc_data['days_ahead'], kde=False, binwidth=7, color='r', alpha=0.4, label ='Discounted', ax=ax)
sns.histplot(no_disc_data['days_ahead'], kde=False, binwidth=7, color='b', alpha=0.4, label='Not Discoutned', ax=ax)
ax.set_xlim(0, 200)
ax.legend()
ax.set_xlabel('Days Ahead')
plt.show()
```

```python
did_not_cancel = desc_data[desc_data['discounted'] == 1]['Cancelled At'].isna().sum() / desc_data[desc_data['discounted'] == 1].shape[0]
print(f'Percentage of discounted reservations that did not cancel: {did_not_cancel * 100:.2f}%')
```

```python
did_not_cancel = desc_data[desc_data['discounted'] == 0]['Cancelled At'].isna().sum() / desc_data[desc_data['discounted'] == 0].shape[0]
print(f'Percentage of non-discounted reservations that did not cancel: {did_not_cancel * 100:.2f}%')
```

```python
import re
from unidecode import unidecode

text = "Apartamento com 2 quartos (4 Adultos) - Rua dos Remédios, n.º 30 - High Season 3 Nights"
n_rooms = re.findall(r'(\d+)\s+quarto', text)
```

```python
def add_n_rooms(text):
    n_rooms = re.findall(r'(\d+)\s+quarto', text)
    if len(n_rooms) > 0:
        return n_rooms[0]
    else:
        return 1
```

```python
def clean_text(text):
    text = unidecode(text)
    text = text.lower()
    return text
```

```python
add_n_rooms(clean_text(text))
```
